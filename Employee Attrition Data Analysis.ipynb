## 6. Exploratory Data Analysis

###  What is Average age of the job attrition person?

### What is the highest qualifacion of attrition employee?

```python
def education(n):
    if n==1:
        return 'Below College'
    elif n==2:
        return 'College'
    elif n==3:
        return 'Bachelor'
    elif n==4:
        return  'Master'
    else:
        return 'Doctor'
df["Highest_Qualification"]=df["Education"].apply(education)
```

```python
sns.countplot(data=df,x="Highest_Qualification",hue="Attrition")
plt.title("Qalification wise attrition")
plt.show()
```

**Observation:**
- **Bachelor,Master and college** pass eomloyee are maximum attrition.
- Doctored degree qualified employee job resign is very low.

### What is the distribution of jobrole wise attrition?

**Observation:**

### Word Cloud

```python
# Create a figure with 12x12 inch dimension
fig = plt.figure(figsize=(12, 12))
stopwords = set(list(STOPWORDS))
stopwords.add('Sales Executive')
stopwords.add('Research Scientist')
stopwords.add('Laboratory Technician')
stopwords.add('Manufacturing Director')
stopwords.add('Healthcare Representative')
stopwords.add('Manager')
stopwords.add('Sales Representative')
stopwords.add('Research Director')
stopwords.add('Human Resources')


wordcloud = WordCloud(background_color = 'darkorchid',
                      stopwords = stopwords, width = 1200, height = 800).generate(str(df["JobRole"]))

plt.title('Common JobRole in Company', fontsize = 18)
print(wordcloud)
plt.axis('off')
plt.imshow(wordcloud)
plt.show()
```

### What is the job attrition ditribution with respect to EnvironmentSatisfaction,JobSatisfaction,RelationshipSatisfaction and WorkLifeBalance?

```python
fig,ax=plt.subplots(2,2,figsize=(10,10))
sns.kdeplot(data=df,x="EnvironmentSatisfaction",hue="Attrition",ax=ax[0][0])
sns.kdeplot(data=df,x="JobSatisfaction",hue="Attrition",ax=ax[0][1])
sns.kdeplot(data=df,x="RelationshipSatisfaction",hue="Attrition",ax=ax[1][0])
sns.kdeplot(data=df,x="WorkLifeBalance",hue="Attrition",ax=ax[1][1])
plt.show()
```

```python
def attrition_mark(n):
    if n=="Yes":
        return 1
    else:
        return 0
df["Attrition_Mark"]=df["Attrition"].apply(attrition_mark)
```

```python
sns.heatmap(df[["Attrition_Mark",'EnvironmentSatisfaction','JobSatisfaction','RelationshipSatisfaction',
                    'WorkLifeBalance']].corr(),annot=True)
plt.title("Co-relation")
plt.show()
```

**Observation:**
- we have see  their is no clear relation between job attrition  with  EnvironmentSatisfaction,JobSatisfaction,RelationshipSatisfaction and WorkLifeBalance

## 7.2 Actionable Insights : 
---

 Based on the provided insights, here are some actionable recommendations for addressing employee attrition:

- **Travel Frequency:**
  Encourage and support employees who travel frequently by providing flexible work arrangements, better travel accommodations,     or incentives to reduce their likelihood of leaving the job.
- **Distance from Home:**
  Investigate the reasons behind the high attrition rate for employees living 2Km to 13Km from work. Consider implementing         measures to make the commute more convenient, such as providing transportation options or remote work opportunities.
- **Departmental Focus:**
  Focus on retention strategies for the Research & Development and Sales departments, as they exhibit higher job attrition.       Conduct surveys or interviews to understand the specific issues in these departments and address them accordingly.
- **Educational Background:**
  Analyze the factors contributing to attrition among Bachelor, Master, and college-educated employees. Consider tailoring         professional development programs, mentorship, or career advancement opportunities based on the educational background to       improve retention.
- **Job Roles:**
  Develop targeted retention plans for job roles with high attrition rates, such as Sales Executives and Laboratory Technicians.   Identify the specific challenges faced by employees in these roles and implement measures to address them.
- **Gender Disparity:**
  Explore the reasons for higher attrition among male employees compared to female employees. Implement gender-sensitive           policies, mentorship programs, or diversity and inclusion initiatives to create a more inclusive work environment.
- **Marital Status:**
  Understand the factors influencing attrition based on marital status. Tailor benefits or support programs to address the         unique needs of married, divorced, and single employees.
- **Income Levels:**
  Address the high attrition among employees with monthly incomes between 2500 to 7500 by reviewing compensation structures,       offering performance-based bonuses, or conducting market analyses to ensure competitive salaries.
- **Salary Hike Percentage:**
  Reevaluate the salary hike structure and ensure it aligns with industry standards. A fair and competitive salary structure can   positively impact employee satisfaction and retention.
- **Years of Experience:**
  Focus on creating a supportive environment for employees with 1 to 5 years of overall work experience, as they show higher       attrition rates. Implement mentorship programs, training initiatives, and career development opportunities.
- **Role Tenure:**
  Develop strategies to retain employees who have worked in the same role for 0 to 4 years. Consider providing opportunities for   skill development, cross-functional training, or career progression within the same role.
- **Managerial Relationships:**
  Improve support and engagement for employees working with current managers for 0 to 5 years. Offer training for managers to     enhance leadership and communication skills, fostering better relationships with their teams.
  
By addressing these specific points, organizations can create targeted interventions to reduce attrition and enhance overall employee satisfaction and retention.
